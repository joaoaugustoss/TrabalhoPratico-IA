# Balanceamento
Código criado para realizar a etapa de pré-processamento em nossa base de dados e salvá-la em um novo arquivo para reduzir a quantidade de células de códigos em nossos outros notebooks.

# Task: 1
## - Import Libraries

```python
import pandas as pd
from collections import Counter
from sklearn.datasets import make_classification
from imblearn.over_sampling import RandomOverSampler
```

# Task 2:
## - Open DataFrame

```python
train_data = pd.read_csv("/content/Reviews_Classified.csv")
train_data = train_data.drop(columns="Unnamed: 0")
test_data = train_data
```

# Task 3:
## - Balance DataFrame

```python
# summarize class distribution
y = train_data.iloc[:,1].values
print(Counter(y))
# define oversampling strategy
oversample = RandomOverSampler(sampling_strategy='not majority')
# fit and apply the transform
X_over, y_over = oversample.fit_resample(train_data, y)
train_data = X_over
y = y_over
# summarize class distribution
print(Counter(y_over))
```

# Task 4:
## - Save DataFrame in a new .csv file

```python
X_over.to_csv('balanced.csv')
```
